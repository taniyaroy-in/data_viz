## CATEGORICAL PLOT

## 1. Count plot

```python
titanic['alive'].unique()
```

```python
sns.countplot(x='sex', hue='class',data=titanic,order=['female','male'])
```

## 2. Strip plot

```python
sns.stripplot(x='sex',y='total_bill',hue='time',data=tips,jitter=True)
```

```python
sns.swarmplot(x='sex',y='total_bill',hue='time',data=tips) # u can see there is no overlap like stripplot..
# all data points are plotted clearly unlike stripplot
```

```python
sns.violinplot(x='sex',y='total_bill',hue='time',data=tips)
# so this shows the distribution of data and we can infer conclusions from this plot accordingly
```

```python
sns.barplot(x='sex',y='age',hue='class',data=titanic, order=['female','male'], estimator=np.std, palette='Set2')
```

```python
sns.boxplot(x='sex',y='age',hue='class',data=titanic, order=['female','male'])
```

```python
#sns.relplot(x='age',y='fare',data=titanic)          # by default kind is scatter
#sns.relplot(x='age',y='fare',data=titanic, kind='line')
sns.relplot(x='age',y='fare',hue='sex',col='class',row='alive',data=titanic, kind='line',palette='coolwarm')
```
