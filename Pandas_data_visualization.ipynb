
## Pandas inbuilt visualization


```python
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
iris.columns
```

```python
iris['species'].unique()
```


To check the details of the dataset e.g. - columns, data types, recs we use info()

```python
iris.info() 
```

```python
iris.plot(figsize=(12,5), title='Iris dataset')  # it plots line chart by default.. plots all the numerical cols
# this is equivalent to iris.plot.line(figsize=(12,5))
```

```python
## to display only 1 record of particularly 1 row

iris.iloc[0]

## now i want to display only the 1st 4 columns
iris.iloc[0].drop('species')

iris.iloc[0].drop('species').plot.bar()  ## this is the plot of the 1st rec
##iris.iloc[1].drop('species').plot.bar()  ## this is the plot of the 2nd rec & so on...
```

```python
iris.plot.scatter(x='petal_length',y='petal_width')  ## just a scatter plot 
```

```python
## i want to get info on histogram of particular cols
iris['petal_length'].plot.hist()

## now suppose i want to get info on histogram of all cols then 
iris.drop('species', axis=1).plot.hist()  ## OR
iris.drop('species', axis=1).plot(kind ='hist')

## these histograms give the mean value of the sepal_length, width, petal_length, petal_width
```

```python
iris.corr()
```

```python
## lets work on titanic dataset now
titanic.head()
```

```python
titanic['age'].plot(kind='hist')
titanic['pclass'].plot(kind='hist')  ## the plots return the mean value of these columns.. 
```

```python
#now say i want to get the count of a categorical col like alone
titanic['alone'].value_counts().plot(kind='barh') 
#titanic.alone.value_counts().plot(kind='bar')  
```

Now i want to plot the kernal density graphs of suppose fare and age columns... so here it is how we do it &

Since we want to plot multiple cols so we have to pass a list of these cols as below:

```python
titanic[['age','fare']].plot(kind='kde') 
```

now if i want to see how above 2 cols are coorelated through graphs/charts then :

```python
titanic[['age','fare']].plot(kind='scatter', x = 'age', y = 'fare') 
```

thus we see that fare and age are not coorelated through above plot.. we can even check it mathematically

```python
titanic[['age','fare']].corr()
```

so we can see 0.09.... almost nothing..
