```python
plt.subplot(2,1,1)  # so here 2 is the no of rows and 1 is no of cols, so the syntax is subplot(no of rows, no of cols, position of the plot).. it is like a matrix..
plt.plot(y,x, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('1st row 1st column')

plt.subplot(2,1,2)
plt.plot(x,a, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2nd row 1st column')

plt.tight_layout()
```
